```python
base_url = "https://www.nykaa.com/skin/sun-care/c/8428?page_no={}&sort=popularity&search_redirection=True&eq=desktop"

# Initialize lists to store the data
product_urls = []
ratings_data = []

# Loop through pages 1 to 5
for page_no in range(1, 6):
    # Construct the URL for the current page
    url = base_url.format(page_no)
    driver.get(url)
    print(f"Fetching data from: {url}")
    
    # Get list of product elements
    product_elements = driver.find_elements(By.CLASS_NAME, "productWrapper.css-17nge1h")
    
    # Loop through each product element to fetch the product URL
    for element in product_elements:
        try:
            product_url_element = element.find_element(By.TAG_NAME, "a")
            product_url = product_url_element.get_attribute("href")
            product_urls.append(product_url)
        except Exception as e:
            print(f"Error: {e}")

# Iterate through the product URLs, open each URL, and fetch the rating
for i, product_url in enumerate(product_urls):
    driver.get(product_url)
    
    try:
        rating_element = driver.find_element(By.CLASS_NAME, "css-m6n3ou")
        rating = rating_element.text
        ratings_data.append([rating])

        print(f"Rating for {product_url}: {rating}")
    except Exception as e:
        print(f"Error fetching rating for {product_url}: {e}")
ratings_df = pd.DataFrame(ratings_data, columns=['Rating'])

        
```
